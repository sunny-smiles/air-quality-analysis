## Exploratory Data Analysis (EDA)

### Concat Data

```python
#station identifier
aotizhongxin_df['station'] = 'Aotizhongxin'
changping_df['station'] = 'Changping'
dingling_df['station']= 'Dingling'
dongsi_df['station'] = 'Dongsi'
guanyuan_df['station'] = 'Guanyuan'
huairou_df['station'] = 'Huairou'
nongzhanguan_df['station'] = 'Nongzhanguan'
shunyi_df['station'] = 'Shunyi'
tiantan_df['station'] = 'Tiantan'
wanliu_df['station'] = 'Wanliu'
wanshouxigong_df['station'] = 'Wanshouxigong'
```

```python
#concat dengan mengabaikan index agar tidak menumpuk
main_data = pd.concat([
    aotizhongxin_df,
    changping_df,
    dingling_df,
    dongsi_df,
    guanyuan_df,
    huairou_df,
    nongzhanguan_df,
    shunyi_df,
    tiantan_df,
    wanliu_df,
    wanshouxigong_df
], axis=0, ignore_index=True)
```

```python
main_data
```

```python
#melihat apakah overlap untuk kolom 'No'
main_data[main_data.No == main_data.No.max()]
```

```python
#agar tidak overlap
main_data.drop(columns=['No'], inplace=True)
main_data
```

### Explore Regional Polution Comparison

```python
#rata-rata polusi tiap tempat
regional_pollution = main_data.groupby('station')[['PM2.5', 'PM10']].mean()
regional_pollution
```

```python
regional_pollution_sorted = regional_pollution.sort_values(by='PM2.5', ascending=False)
print("Average Pollution Levels by Station:\n", regional_pollution_sorted)
```

### Explore Seasonal Changes in Air Quality

```python
#map months to season
def month_to_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
```

```python
main_data['season'] = main_data['month'].apply(month_to_season)
```

```python
seasonal_pollution = main_data.groupby('season')[['PM2.5', 'PM10']].mean()

print("Average Pollution Levels by Season:\n", seasonal_pollution)
```

```python
monthly_pollution = main_data.groupby('month')[['PM2.5', 'PM10']].mean()

print("Average Pollution Levels by Month:\n", monthly_pollution)
```

### Explore Effect of Wind on Pollution Levels

```python
wind_direction_pollution = main_data.groupby('wd')[['PM2.5', 'PM10']].mean()

print("Average Pollution Levels by Wind Direction:\n", wind_direction_pollution)
```

```python
wind_speed_pollution = main_data.groupby('WSPM')[['PM2.5', 'PM10']].mean()

print("Average Pollution Levels by Wind Speed:\n", wind_speed_pollution)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana tingkat polusi (PM2.5, PM10) dibandingkan di antara tempat pemantauan yang berbeda? Tempat mana yang secara konsisten melaporkan polusi yang lebih tinggi, dan dapatkah hal ini mengindikasikan area yang lebih tercemar?

```python
#menampilkan boxplot PM2.5 tiap tempat
plt.figure(figsize=(14, 7))
sns.boxplot(data=main_data, x='station', y='PM2.5')
plt.title('Comparison of PM2.5 Levels Across Stations')
plt.xlabel('Station')
plt.ylabel('PM2.5 Concentration')
plt.xticks(rotation=90)
plt.show()
```

```python
#menampilkan boxplot PM10 tiap tempat
plt.figure(figsize=(14, 7))
sns.boxplot(data=main_data, x='station', y='PM10')
plt.title('Comparison of PM10 Levels Across Stations')
plt.xlabel('Station')
plt.ylabel('PM10 Concentration')
plt.xticks(rotation=90)
plt.show()
```

### Pertanyaan 2: Bagaimana kualitas udara (berdasarkan tingkat PM2.5 dan PM10) berubah di berbagai bulan dan musim? Apakah ada waktu-waktu tertentu dalam setahun ketika polusi udara lebih buruk?

```python
#rata-rata polusi perbulan
monthly_avg = main_data.groupby('month')[['PM2.5', 'PM10']].mean()
```

```python
#menampilkan line chart untuk melihat pola
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg.index, monthly_avg['PM2.5'], marker='o', label='PM2.5')
plt.plot(monthly_avg.index, monthly_avg['PM10'], marker='o', label='PM10')
plt.title('Seasonal Variation in Air Quality')
plt.xlabel('Month')
plt.ylabel('Average Concentration')
plt.legend()
plt.show()
```

### Pertanyaan 3: Bagaimana kecepatan dan arah angin (WSPM) berdampak pada tingkat polusi (PM2.5, PM10) di berbagai tempat? Apakah angin yang lebih kencang membantu mengurangi polusi, atau justru membawa lebih banyak polutan dari daerah lain?

```python
#menampilkan scatterplot untuk melihat efek angin terhadap PM2.5
plt.figure(figsize=(14, 7))
sns.scatterplot(data=main_data, x='WSPM', y='PM2.5', hue='wd', palette='coolwarm')
plt.title('Effect of Wind Speed and Direction on PM2.5 Levels')
plt.xlabel('Wind Speed (WSPM)')
plt.ylabel('PM2.5 Concentration')
plt.show()
```

```python
#menampilkan scatterplot untuk melihat efek angin terhadap PM10
plt.figure(figsize=(14, 7))
sns.scatterplot(data=main_data, x='WSPM', y='PM10', hue='wd', palette='coolwarm')
plt.title('Effect of Wind Speed and Direction on PM10 Levels')
plt.xlabel('Wind Speed (WSPM)')
plt.ylabel('PM10 Concentration')
plt.show()
```

## Conclusion

- Conclusion pertanyaan 1:
<b>Aotizhongxin, Dongsi, dan Tiantan</b> adalah tempat-tempat yang secara konsisten melaporkan tingkat polusi yang lebih tinggi, terutama dalam hal frekuensi dan besarnya pencilan. Tempat-tempat ini menunjukkan area yang mungkin lebih rentan terhadap kejadian polusi tinggi, yang menunjukkan bahwa tempat-tempat ini terletak di wilayah yang lebih tercemar atau lebih sering terpengaruh oleh sumber polusi dibandingkan tempat lainnya.

- Conclution pertanyaan 2:
Kualitas udara cenderung <b>memburuk selama musim dingin</b>, dengan tingkat polusi yang lebih tinggi, menunjukkan perlunya intervensi yang lebih intensif selama bulan-bulan ini.

- Conclusion pertanyaan 3:
Kecepatan angin yang lebih tinggi secara umum efektif dalam mengurangi konsentrasi kedua jenis polutan tersebut, dengan tren penurunan yang jelas pada konsentrasi PM2.5 dan PM10 seiring peningkatan kecepatan angin. Arah angin tertentu, seperti dari arah S dan SW, cenderung terkait dengan konsentrasi polutan yang lebih tinggi pada kecepatan angin rendah, mengindikasikan bahwa angin dari arah ini mungkin membawa polutan dari area yang lebih tercemar. Secara keseluruhan, kecepatan angin memainkan peran lebih signifikan dalam menyebarkan dan mengurangi konsentrasi polutan, meskipun arah angin juga berkontribusi pada variasi lokal dalam tingkat polusi udara.
